# Blending and Pasting Images

Blending and pasting images in image processing involves combining two or more images together to create a single, composite image. Blending typically involves overlaying one image on top of another, adjusting the transparency of the top image to create a seamless transition between the two. Pasting involves taking a section of one image and placing it onto another image, potentially adjusting the size and position of the pasted image to fit seamlessly with the background image.

## Blending the Image

We will blend the values together with the formula:

###  $$ img1 * \alpha  + img2 * \beta  + \gamma $$

```python
blended = cv2.addWeighted(src1=img1,alpha=0.7,src2=img2,beta=0.3,gamma=0)
```

The above line of code is to blend two images, img1 and img2, together. The blending is done by adjusting the transparency of the two images and adding them together.

The alpha value of 0.7 means that the first image, img1, will be more visible in the final blended image, while the beta value of 0.3 means that the second image, img2, will be less visible. The gamma value of 0 is a bias term that is added to the final blended image. You can play with the values of Aplha and Beta as per your requirement

So, in laymen's terms, this code is taking two images, img1 and img2 and blending them together in a way that img1 will be more visible than img2 in the final blended image. The final blended image will be a combination of both images, with img1 being more prominent. This is done by adjusting the transparency of the images in a specific ratio and adding them together.

```python
plt.imshow(blended)
```

## Overlaying Images of Different Sizes

We can use this quick trick to quickly overlap different sized images, by simply reassigning the larger image's values to match the smaller image.

```python
img1 = cv2.imread('../Data/mona_lisa.jpeg')
img2 = cv2.imread('../Data/watermark.jpeg')
img2 =cv2.resize(img2,(600,600))

img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)

large_img = img1
small_img = img2
```

```python
x_offset=0
y_offset=0
```

```python
large_img[y_offset:y_offset+small_img.shape[0], x_offset:x_offset+small_img.shape[1]] = small_img
```

```python
plt.imshow(large_img)
```

## Blending Images of Different Sizes

```python
img1 = cv2.imread('../Data/mona_lisa.jpeg')
img2 = cv2.imread('../Data/watermark.jpeg')
img2 =cv2.resize(img2,(600,600))

img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

### Create a Region of Interest (ROI)

```python
img1.shape
```

```python
x_offset=800-600
y_offset=1192-600
```

```python
# Creating an ROI of the same size of the foreground image (smaller image that will go on top)
rows,cols,channels = img2.shape
# roi = img1[0:rows, 0:cols ] # TOP LEFT CORNER
roi = img1[y_offset:1192,x_offset:800] # BOTTOM RIGHT CORNER
```

```python
plt.imshow(roi)
```

```python
roi.shape
```

```python
# Now create a mask of logo and create its inverse mask also
img2gray = cv2.cvtColor(img2,cv2.COLOR_BGR2GRAY)
```

```python
img2gray.shape 
```

```python
plt.imshow(img2gray,cmap='gray')
```

```python
mask_inv = cv2.bitwise_not(img2gray)
ret,thresh1 = cv2.threshold(mask_inv,127,255,cv2.THRESH_BINARY)
```

```python
thresh1.shape
```

```python
plt.imshow(thresh1,cmap='gray')
```

## Convert Mask to have 3 channels

```python
white_background = np.full(img2.shape, 255, dtype=np.uint8)
bk = cv2.bitwise_or(white_background, white_background, mask=thresh1)
```

```python
bk.shape
```

```python
plt.imshow(bk)
```

```python
fg = cv2.bitwise_or(img2, img2, mask=thresh1)
```

```python
plt.imshow(fg)
```

```python
fg.shape
```

### Get ROI and blend in the mask with the ROI

```python
final_roi = cv2.bitwise_or(roi,fg)
```

```python
plt.imshow(final_roi)
```

### Now add in the rest of the image

```python
large_img = img1
small_img = final_roi


large_img[y_offset:y_offset+small_img.shape[0], x_offset:x_offset+small_img.shape[1]] = small_img

plt.imshow(large_img)
```
