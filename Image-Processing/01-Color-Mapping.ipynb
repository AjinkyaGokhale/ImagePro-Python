# Colorspaces

Refer this link for more details: https://docs.opencv.org/4.x/df/d9d/tutorial_py_colorspaces.html

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
img = cv2.imread('../Data/tiger.jpeg')
duplicate_img =img
```

```python
plt.imshow(img)
```

**Matplotlib --> RGB Scale but OpenCV --> BGR Scale**

Hence Image appered seems wrongly colored. 
So, We will change the colorSpace

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

## Converting to HSV
HSV (Hue-Saturation-Value) color space is a cylindrical representation of the RGB color space, where the hue component represents the dominant wavelength of light, the saturation component represents the purity of the light, and the value component represents the brightness of the light.

```python
duplicate_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.imshow(duplicate_img)
```

## Converting to HLS Color space.
The HLS (Hue-Lightness-Saturation) color space is a cylindrical representation of the RGB color space, where the hue component represents the dominant wavelength of light, the lightness component represents the intensity of the light, and the saturation component represents the purity of the light.

```python
img = cv2.imread('../Data/tiger.jpeg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2HLS)
plt.imshow(img)
```
